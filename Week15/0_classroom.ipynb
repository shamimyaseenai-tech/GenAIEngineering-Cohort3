```python
import random
import numpy as np
import matplotlib.pyplot as plt
```

```python

# Method 1: Using random.uniform()
def generate_ages_random(n, min_age=0, max_age=100):
    """Generate n random ages using random.uniform()"""
    ages = [random.uniform(min_age, max_age) for _ in range(n)]
    return ages

ages1 = generate_ages_random(100, 18, 60)
print(ages1)
plt.figure(figsize=(10, 6))
plt.hist(ages1, range(len(ages1)))
plt.xlabel('Age')
plt.ylabel('Index')
plt.title('Age Bar Chart')
plt.show()
```

```python
# Method 2: Using numpy.random.uniform()
def generate_ages_numpy(n, min_age=0, max_age=100):
    """Generate n random ages using numpy"""
    ages = np.random.uniform(min_age, max_age, size=n)
    return ages
ages2 = generate_ages_numpy(10, 18, 65)
print(ages2)
plt.figure(figsize=(10, 6))
plt.bar(range(len(ages2)), ages2)
plt.xlabel('Index')
plt.ylabel('Age')
plt.title('Age Bar Chart')
plt.show()
```

```python

# Method 3: For integer ages only
def generate_integer_ages(n, min_age=0, max_age=100):
    """Generate n random integer ages"""
    ages = [random.randint(min_age, max_age) for _ in range(n)]
    return ages
ages3 = generate_integer_ages(10, 18, 65)
print(ages3)

plt.figure(figsize=(10, 6))
plt.bar(range(len(ages3)), ages3)
plt.xlabel('Index')
plt.ylabel('Age')
plt.title('Age Bar Chart')
plt.show()
```

```python
# For realistic adult ages
adult_ages = np.random.uniform(18, 80, size=2000)
print("Adult ages:", [round(age) for age in adult_ages])

# Create histogram
plt.figure(figsize=(10, 6))
plt.hist(adult_ages, bins=20, edgecolor='black', alpha=0.7)
plt.xlabel('Age')
plt.ylabel('Frequency')
plt.title('Distribution of Random Ages')
plt.grid(axis='y', alpha=0.3)
plt.show()
```

```python
from scipy import stats
# Create a visual comparison of discrete vs continuous distributions
fig, axes = plt.subplots(1, 2, figsize=(16, 6))
fig.suptitle('Discrete vs Continuous Distributions', fontsize=16, fontweight='bold')

# Discrete Distribution Example: Binomial
n, p = 10, 0.3
x_discrete = np.arange(0, n+1)
y_discrete = stats.binom.pmf(x_discrete, n, p)

print(np.round(y_discrete, 3))  # Round for better readability

axes[0].bar(x_discrete, y_discrete, alpha=0.7, color='steelblue', edgecolor='black')
axes[0].set_title('Discrete Distribution\n(Binomial: n=10, p=0.3)', fontweight='bold')
axes[0].set_xlabel('Number of Successes (k)')
axes[0].set_ylabel('Probability P(X = k)')
axes[0].grid(True, alpha=0.3)
axes[0].set_xticks(x_discrete)
```
